```python
def convert_fee_to_numeric(fee_str):
    # Usunięcie znaków specjalnyc
    fee_str = fee_str.replace('€', '').replace('m', 'e6').replace('k', 'e3')
    return float(fee_str)
```

```python
import numpy as np
def delete_strings(row):
    if isinstance(row, str):
        if ('loan' in row) or ('fee' in row) or ('?' in row) or ('draft' in row) or ('free' in row):
            row= np.nan
            return row
        elif row.endswith('m'):
            row= row.replace('m', '')
            row= row.replace('€', '')
            row = float(row)*1e6
            return row
        elif row.endswith('k'):
            row= row.replace('k', '')
            row= row.replace('€', '')
            row = float(row)*1e3
            return row

    return row



players_df['fee']= players_df['fee'].apply(delete_strings)
```
